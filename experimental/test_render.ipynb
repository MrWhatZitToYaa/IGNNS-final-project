```python
from bcnf.simulation.render import render
```

```python
data = render('preliminary_data_4s.pkl', n = 10, T = 4, FPS = 30)
```

```python
cams = data['cams']
```

```python
cam1 = cams[6][0]
cam2 = cams[6][1]
```

```python
# make gif of cam 1 and cam 2

import matplotlib.pyplot as plt
import numpy as np
import matplotlib.animation as animation
from bcnf.utils import get_dir


aspect_ratio = 16 / 9

fov_vertical = 70.0 / aspect_ratio

phi = (70.0 / 2) * (np.pi / 180)
theta = (fov_vertical / 2) * (np.pi / 180)


fig, ax = plt.subplots(figsize=(16, 9))
ax.set_xlabel('horizontal angle')
ax.set_ylabel('vertical angle')
ax.title.set_text('cam1')
gif = []
for i in range(len(cam1)):
    gif.append([ax.imshow(cam1[i], extent=[-phi, phi, -theta, theta], cmap='hot', animated=True)])

ani = animation.ArtistAnimation(fig, gif, interval=33, blit=True, repeat_delay=3000)
ani.save(f'{get_dir()}/tests/gifs/cam1.gif', writer='imagemagick')
```

```python
fig, ax = plt.subplots(figsize=(16, 9))
ax.set_xlabel('horizontal angle')
ax.set_ylabel('vertical angle')
ax.title.set_text('cam2')
gif = []
for i in range(len(cam1)):
    gif.append([ax.imshow(cam2[i], extent=[-phi, phi, -theta, theta], cmap='hot', animated=True)])

ani = animation.ArtistAnimation(fig, gif, interval=33, blit=True, repeat_delay=3000)
ani.save(f'{get_dir()}/tests/gifs/cam2.gif', writer='imagemagick')
```

```python
# visualise first image of cam 1

fig, ax = plt.subplots(figsize=(16, 9))
ax.set_xlabel('horizontal angle')
ax.set_ylabel('vertical angle')
ax.title.set_text('cam1')
ax.imshow(cam1[3], extent=[-phi, phi, -theta, theta], cmap='hot')
plt.show()
```

```python
traj = data['traj']
trajectory = traj[9]

# visualise 3d trajectory

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.title.set_text('trajectory')
ax.plot(trajectory[:, 0], trajectory[:, 1], trajectory[:, 2])
plt.show()
```
